### Creating dataset for training with GPT-4 training data

```python
import pandas as pd
```

```python
#loading in exisitng data

reviews_df = pd.read_csv('fake reviews dataset.csv')

reviews_df['labels'] = [1 if label=='OR' else 0 for label in reviews_df['label']]


# how many fake reviews are there?
print('Number of fake reviews before removal:', len(reviews_df[reviews_df['labels']==1]))

# remove 500 fake reviews
reviews_df = pd.concat([reviews_df[reviews_df['labels']==1].iloc[150:], reviews_df[reviews_df['labels']==0]])

# how many fake reviews are there?
print('Number of fake reviews after removal :', len(reviews_df[reviews_df['labels']==1]))
```

```python
# read text file 
gpt_reviews = infile = open('final_GPT_4_reviews.csv', 'r').read().split('\n')

# create dataframe
gpt_reviews_df = pd.DataFrame({'text_':gpt_reviews, 'labels':[0]*len(gpt_reviews)})


reviews_df = pd.concat([reviews_df, gpt_reviews_df])


# shuffle to avoid bias
reviews_df = reviews_df.sample(frac=1).reset_index(drop=True)


reviews_df.to_csv('final_reviews.csv', index=False)
```
