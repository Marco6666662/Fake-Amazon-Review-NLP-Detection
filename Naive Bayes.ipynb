```python
# imports necessary libraries/packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import cross_val_score, KFold, cross_val_predict
from sklearn.feature_extraction.text import TfidfVectorizer
```

## Real/Fake Amazon Reviews Data Loading/Cleaning

```python
# sets the name of the file
FILENAME = 'fake reviews dataset.csv'

# reads the csv file into a pandas df
df = pd.read_csv(FILENAME)
df
```

```python
# creates a column for the labels (0 if the review is fake (CG) and 1 if the review is real (OR))
df['labels'] = [1 if label=='OR' else 0 for label in df['label']]
df
```

## Naive Bayes Model on original real/fake Amazon reviews

```python
# splits the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df['text_'], df['labels'], test_size=0.2, random_state=42)
```

```python
# converts the text data into numerical features using CountVectorizer
vectorizer = CountVectorizer()
train_features = vectorizer.fit_transform(X_train)
test_features = vectorizer.transform(X_test)

test_features
```

```python
# trains the Naive Bayes classifier
naive_bayes_classifier = MultinomialNB()
naive_bayes_classifier.fit(train_features, y_train)
```

```python
# makes predictions on the test set
predictions = naive_bayes_classifier.predict(test_features)

# displays the full content of the predictions
np.set_printoptions(threshold=np.inf)

print(predictions)
```

```python
# evaluates the performance metrics
accuracy = accuracy_score(y_test, predictions)
confusion_mat = confusion_matrix(y_test, predictions)
classification_rep = classification_report(y_test, predictions)

print('accuracy:', accuracy)
print('confusion_matrix:', confusion_mat)
print('classification_report:', classification_rep)
```

```python
# plots the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_mat, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix for Naive Bayes Classifier (Train/Test Split)')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

## Naive Bayes Model on Chat GPT-4 Data

```python
# loads the chat gpt 4 generated reviews
gpt4_df = pd.read_csv('generated_amazon_reviews.csv')
gpt4_df['labels'] = [0] * len(gpt4_df.Review)
gpt4_df
```

```python
# converts the gpt 4 text data into numerical features using CountVectorizer
gpt4_test_features = vectorizer.transform(gpt4_df['Review'])

gpt4_test_features
```

```python
# makes predictions on the gpt 4 test set
gpt4_predictions = naive_bayes_classifier.predict(gpt4_test_features)

# displays the full content of the predictions
np.set_printoptions(threshold=np.inf)

print(gpt4_predictions)
```

```python
# evaluates the performance metrics
accuracy = accuracy_score(gpt4_df['labels'], gpt4_predictions)
confusion_mat = confusion_matrix(gpt4_df['labels'], gpt4_predictions)
classification_rep = classification_report(gpt4_df['labels'], gpt4_predictions)

print('accuracy:', accuracy)
print('confusion_matrix:', confusion_mat)
print('classification_report:', classification_rep)
```

## K-Fold Cross Validation on original data

```python
# converts the text data to numerical features using TF-IDF
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df['text_'])

n_splits = 5

# sets up k-fold cross-validation
kfold = KFold(n_splits=n_splits, shuffle=True, random_state=42)

# performs k-fold cross-validation and gets the accuracy scores for it
cross_val_scores = cross_val_score(naive_bayes_classifier, X, df['labels'], cv=kfold, scoring='accuracy')

# displays the accuracy scores for each fold
print("Cross-validation Scores:", cross_val_scores)

# calculates and displays the mean accuracy across all folds
mean_accuracy = cross_val_scores.mean()
print("Mean Accuracy:", mean_accuracy)
```

```python
# plots the cross-validation scores as a line graph
plt.figure(figsize=(8, 6))
plt.plot(range(1, n_splits + 1), cross_val_scores, marker='o', linestyle='-')
plt.xlabel('Fold')
plt.ylabel('Accuracy')
plt.title('Cross-validation Scores for Naive Bayes Classifier')
plt.ylim(0.8, 0.9)
plt.grid(True)
plt.show()
```

```python
# performs k-fold cross-validation and gets predictions
predictions = cross_val_predict(naive_bayes_classifier, X, df['labels'], cv=kfold)

# displays the confusion matrix for each fold
plt.figure(figsize=(15, n_splits))

for i, (train_index, test_index) in enumerate(kfold.split(X)):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = df['labels'].iloc[train_index], df['labels'].iloc[test_index]
    
    naive_bayes_classifier.fit(X_train, y_train)
    fold_predictions = naive_bayes_classifier.predict(X_test)
    
    confusion_mat = confusion_matrix(y_test, fold_predictions)
    
    plt.subplot(2, 5, i+1)
    sns.heatmap(confusion_mat, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(f'Fold {i+1}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')

plt.suptitle('Confusion Matrices for Each Fold', y=1.02)
plt.tight_layout()
plt.show()
```
